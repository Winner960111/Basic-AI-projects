# Heart Disease Detection

### Import dependencies

```python
import numpy as pandas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Data Collection and Processing

```python
# loading dataset
heart_dataset = pd.read_csv("heart_disease_data.csv")
```

```python
# showing dataset
heart_dataset.head()
```

```python
# Create multiple plots
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns of subplots

# Plot on the first subplot
axs[0].plot([1, 2, 3, 4], [1, 4, 9, 16])
axs[0].set_title('Plot 1')

# Plot on the second subplot
axs[1].plot([1, 2, 3, 4], [4, 3, 2, 1])
axs[1].set_title('Plot 2')

plt.show()
```

```python
num_rows, num_cols = 1, len(heart_dataset.columns)
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 5))

col = heart_dataset["age"]
sns.countplot(x = col, ax = axs[0])
axs[0].set_title("Count Plot for age")

plt.tight_layout()
plt.show()
```

```python
num_rows, num_cols = 2, 5
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 5))

for index, col_name in enumerate(heart_dataset.columns):
    if(col_name != "trestbps" and col_name != "chol" and col_name != "thalach" and col_name != "oldpeak"):
        col = heart_dataset[col_name]
        sns.countplot(x = col, ax = axs[index])
        axs[index].set_title("Count Plot for ")

plt.tight_layout()
plt.show()
```

```python
# show counts
for i in heart_dataset.columns:
    if(i != "trestbps" and i != "chol" and i != "thalach" and i != "oldpeak"):
        plt.figure(figsize=(8,8))
        col = heart_dataset[i]
        sns.countplot(x = col)
        plt.show()
```
